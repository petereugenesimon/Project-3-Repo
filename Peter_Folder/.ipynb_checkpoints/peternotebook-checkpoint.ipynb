```python
cod_file = "pes_grouped_data_deleted_rows.csv"
```

```python
cod_df = pd.read_csv(cod_file)
```

```python
cod_df.head()
```

```python
reduced_cod_df = cod_df[["MMWR Year", "Jurisdiction of Occurrence", "Natural Cause", "Diabetes mellitus (E10-E14)", "Alzheimer disease (G30)", "Influenza and pneumonia (J09-J18)", "Symptoms, signs and abnormal clinical and laboratory findings, not elsewhere classified (R00-R99)", "Diseases of heart (I00-I09,I11,I13,I20-I51)", "COVID-19 (U071, Multiple Cause of Death)", "COVID-19 (U071, Underlying Cause of Death)"]]
reduced_cod_df.head()
```

```python
renamed_cod_df = reduced_cod_df.rename(columns={"MMWR Year": 'Year', "Jurisdiction of Occurrence": 'State', "Diabetes mellitus (E10-E14)": 'Diabetes', "Alzheimer disease (G30)": 'Alzheimers', "Influenza and pneumonia (J09-J18)": 'Influenza_pneumonia', "Symptoms, signs and abnormal clinical and laboratory findings, not elsewhere classified (R00-R99)": 'Symptoms_not_elsewhere_classified', "Diseases of heart (I00-I09,I11,I13,I20-I51)": 'Diseases_of_heart', "COVID-19 (U071, Multiple Cause of Death)": 'COVID_19_Multiple', "COVID-19 (U071, Underlying Cause of Death)": 'COVID_19_Underlying'})
renamed_cod_df.head()
```
