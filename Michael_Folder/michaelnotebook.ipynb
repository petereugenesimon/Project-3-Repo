```python
import sqlite3
import csv
```

```python
def create_sqlite_from_csv(csv_file, sqlite_file, table_name):
    # Connect to the SQLite database
    conn = sqlite3.connect(sqlite_file)
    cursor = conn.cursor()
    
    # Create the table in the SQLite database
    create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} (id INTEGER, year TEXT, state TEXT, all_causes INTEGER, natural_cause INTEGER, septicemia INTEGER, malignant_neoplasms INTEGER, diabetes INTEGER, alzheimers INTEGER, influenza_pneumonia INTEGER, lower_respiratory INTEGER, other_respiratory INTEGER, nephritis INTEGER, abnormal INTEGER, heart_disease INTEGER, cerebrovascular_diseases INTEGER, covid INTEGER)"
    cursor.execute(create_table_query)

    # Read the CSV file and insert the data into the SQLite table
    with open(csv_file, 'r') as file:
        csv_data = csv.reader(file)
        next(csv_data)  # Skip the header row
        
        insert_query = f"INSERT INTO {table_name} (id, year, state, all_causes, natural_cause, septicemia, malignant_neoplasms, diabetes, alzheimers, influenza_pneumonia, lower_respiratory, other_respiratory, nephritis, abnormal, heart_disease, cerebrovascular_diseases, covid) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)"
        
        for row in csv_data:
            cursor.execute(insert_query, row)
    
    # Commit the changes and close the connection
    conn.commit()
    conn.close()
```

```python
# Specify the paths and names of the CSV and SQLite files
csv_file_path = 'grouped_data_deleted_rows.csv'
sqlite_file_path = 'deathStats.sqlite'
table_name = 'deathStats'
# Call the function to create the SQLite file from the CSV
create_sqlite_from_csv(csv_file_path, sqlite_file_path, table_name)
```
